### First, let's import all of our requirements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import datetime as dt
```

### Now, let's import our data and do some data cleaning

```python
# Electricity prices from http://ets.aeso.ca/
pool_price = pd.read_csv('data/pool_price.csv')
# Crude and INDU prices from Bloomberg
crude_price = pd.read_csv('data/CL1_price.csv')
index_price = pd.read_csv('data/SPY_price.csv')
```

AESO uses an odd 24th hour instead of a 0th hour.

```python
# This function will convert the dates properly, note it uses Python
# 3.5+ static type hints. To run on earlier Python versions, the
# function declaration should look like: def aeso_date(date):
def aeso_date(date: str) -> dt.datetime:
    if date[11:13] != '24':
        return pd.to_datetime(date, format='%m/%d/%Y %H')
    else:
        date = date[0:11] + '00'
        return pd.to_datetime(date, format='%m/%d/%Y %H') + dt.timedelta(days=1)


# First we have to drop the redundant columns with the '*' character
pool_price = pool_price[~pool_price['Date'].str.contains('\*')]
# Now let's convert the date strings to datetimes
pool_price['Date'] = pool_price['Date'].apply(aeso_date)
# Finally, let's only keep the data we need
pool_price.set_index('Date', inplace=True)
pool_price.drop(['30Ravg', 'AIL Demand'], axis=1, inplace=True)
```

Now let's deal with the Bloomberg data.

```python
# Again we've got static typing hints that requires Python 3.5+
def bloomberg_data(df: pd.DataFrame) -> pd.DataFrame:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df.rename(columns={'LAST_PRICE': 'Price'}, inplace=True)
    return df


crude_price = bloomberg_data(crude_price)
index_price = bloomberg_data(index_price)
```

### Let's do some exploratory data analysis

```python
# Let's start with a plotting the raw data
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Pool Price')
ax.plot(pool_price)
fig.autofmt_xdate()
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Crude Price')
ax.plot(crude_price)
fig.autofmt_xdate()
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Index Price')
ax.plot(index_price)
fig.autofmt_xdate()
plt.show()
```

Clearly electricity is more volatile than the other two signals. What is remarkable is how much more volatile it is.

One thing that is interesting about electricity is that it is highly cyclical. During a day, the electricity demand is much higher during peak hours than during the middle of the night, thus the price is cyclical within the day. Also, there is more demand in some seasons rather than others. To illustrate this unusual cyclicality, let's plot our three signals over each day, over the total time period.

```python
def cool_cyclical_plot(frame: pd.DataFrame, delete_outliers: int=0, title: str="") -> None:
    # We need to copy the data so we don't mess up our original dataframes
    df = frame.copy(deep=True)
    # If this is the electricity data, cap the hight to make pattern visible
    if delete_outliers == 1:
        a = np.array(df['Price'].values.tolist())
        df.drop('Price', axis=1, inplace=True)
        df['Price'] = np.where(a > 120, 120, a).tolist()
    df.reset_index(inplace=True)
    df['Hour'] = df['Date'].dt.hour
    df['Date'] = df['Date'].dt.date
    df['Days'] = (df['Date'] - df['Date'][0]).dt.days
    hours = len(pd.unique(df['Hour']))
    df = df.pivot(index='Days', columns='Hour', values='Price')
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.set_title(title)
    X = np.asarray(df.index.tolist())
    Y = np.asarray(range(0,hours))
    X, Y = np.meshgrid(X, Y)
    Z = np.asarray(df.values.tolist()).T
    Z = np.nan_to_num(Z)
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                           linewidth=0.1, antialiased=True)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    ax.view_init(30, 240)
    plt.show()
```

```python
cool_cyclical_plot(pool_price, delete_outliers=1, title="Pool Price")
cool_cyclical_plot(crude_price, title="Crude Price")
cool_cyclical_plot(index_price, title="Index Price")
```

We can clearly see the interesting cyclically characteristics of the electricity prices. The crude plot has many holes in it, because it doesn't actually trade 24 hours per day, but rather 23 hours on most business days, 8 hours on Sundays, and some other weird rules. Also, notice that the index plot only contains 7 hours of data, this is because of the short trading hours of equities. 

### Finally, let's calculate volatility

We will calculate weekly volatility, based on the hours in the day.

```python
def vol_calc(frame: pd.DataFrame, sample: str) -> pd.DataFrame:
    df = frame.copy(deep=True)
    # First calculate returns
    df = df.pct_change()
    df = df.fillna(value=0)
    df = df + 1
    # Then calculate the vol. This line is where "the magic" happens.
    df = df.resample(sample).agg(['std', 'count']).dropna()
    # Now let's move our sub-columns out into the main dataframe
    # There's probably an easier way to do this
    df['std'] = df['Price']['std']
    df['count'] = df['Price']['count']
    df.drop('Price', axis=1, inplace=True)
    # Finally we scale the volatility based on the datapoints in the period
    df['vol'] = df['std'] * np.sqrt(df['count'])
    df.drop(['std', 'count'], axis=1, inplace=True)
    return df
```

```python
# By using 'W-MON' we resample into weekly bars starting on Monday's
crude_vol = vol_calc(crude_price, 'W-MON')
index_vol = vol_calc(index_price, 'W-MON')
pool_vol = vol_calc(pool_price, 'W-MON')
```

First let's view the volatility over time of each time series.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Pool Vol')
ax.plot(pool_vol)
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
fig.autofmt_xdate()
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Crude Vol')
ax.plot(crude_vol)
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
fig.autofmt_xdate()
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Index Vol')
ax.plot(index_vol)
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
fig.autofmt_xdate()
plt.show()
```

It is clear to see that equity volatility is much lower than crude volatility, which is much, much lower than electricity volatility.

### Long Term Data

Now let's look at these trends over the long term. Unfortunately to do this with free data we must use lower resolution data, and only look at crude and the SPY. Here I use ~18 years of daily data from Bloomberg.

```python
index = bloomberg_data(pd.read_csv('data/SPY_long.csv'))
crude = bloomberg_data(pd.read_csv('data/CL1_long.csv'))
```

```python
# Using the 'A' switch, we calculate annual volatility.
index_vol = vol_calc(index, 'A')
crude_vol = vol_calc(crude, 'A')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Crude vs Index Volatility')
ax.plot(crude_vol, label='Crude')
ax.plot(index_vol, label='SPY')
vals = ax.get_yticks()
ax.set_yticklabels(['{:3.2f}%'.format(x*100) for x in vals])
fig.autofmt_xdate()
ax.legend()
plt.show()
```

We can see that crude volatility is significantly higher than index volatility.
